```python
from typing import Dict, Any
import configparser
```

```python
def update_ini_file(ini_file_path: str, update_values: Dict[str, Any]) -> None:
    """
    Update values in an INI file based on a dictionary.
    
    Args:
        ini_file_path: Path to the INI file to update
        update_values: Dictionary with keys in format 'section.key' and corresponding values
                      For example: {'Animal.string_animID': '12345', 'Project.string_date': '2025-08-01'}
    
    Returns:
        None
    """
    # Check if file exists
    if not os.path.exists(ini_file_path):
        raise FileNotFoundError(f"INI file not found: {ini_file_path}")
    
    # Read the INI file
    config = configparser.ConfigParser(allow_no_value=True)
    config.read(ini_file_path)
    
    # Track changes made
    changes_made = []
    
    # Update values from the dictionary
    for key, value in update_values.items():
        # Split the key into section and option
        if '.' in key:
            section, option = key.split('.', 1)
            
            # Check if section and option exist
            if section in config and option in config[section]:
                # Convert value to string based on type
                if value is None:
                    str_value = ""
                elif isinstance(value, bool):
                    str_value = str(value)
                else:
                    str_value = str(value)
                
                # Update the value
                old_value = config[section][option]
                config[section][option] = str_value
                changes_made.append((section, option, old_value, str_value))
            else:
                print(f"Warning: Key '{key}' not found in the INI file")
        else:
            print(f"Warning: Invalid key format '{key}'. Expected format: 'section.option'")
    
    # Write the updated config back to the file
    with open(ini_file_path, 'w') as f:
        config.write(f)
    
    # Print summary of changes
    if changes_made:
        print(f"Updated {len(changes_made)} values in {ini_file_path}:")
        for section, option, old, new in changes_made:
            print(f"  [{section}] {option}: '{old}' → '{new}'")
    else:
        print(f"No changes made to {ini_file_path}")
```
